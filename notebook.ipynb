## Design a Food Rating System 

## Problem-2353



```python
class FoodRatingSystem:
    def __init__(self, foods:List[str],ratings:List[int],cuisine:List[str]):

        self.food_info = { } 
        self.cuisine_info = defaultdict(list)

        for food,cuis,rate in zip(foods,ratings,cuisine):
            self.food_info[food] = (cuis,rate)
            heapq.heappush(self.cuisine_info[cuis],(-rate,food)) ## here we use -rate because, the heapq returns the smallest of the element ,,  but we need the maximum number so


    
    def changeRating(self, food: str, newRating: int) -> None:
        # now we decouple the items 
        cuis, _rate = self.food_info[food]
        self.food_info[food] = (cuis,newRating)
        heapq.heappush(self.cuisine_info[cuis],(-newRating,food))

    
    def highestRated(self, cuisine: str) -> str:
        heap_structure  = self.cuisine_info[cuisine]

        while True:
            rating , food = heap_structure[0]
            _cuisine, _rating = self.food_info[food]

            # valid if rating matches the most recent rating
            if _cuisine == cuisine and rating == _rating:
                return food

            # otherwise outdated entry, discard
            heapq.heappop(heap_structure)
```

### Adding Two Numbers 


```python
class ListNode:
    def __init__(self,val=0,next=None):
        self.val = 0 
        self.next = next 
    

class TwoNumbers:
    def __init__(self,l1:Optional[ListNode], l2:Optional[ListNode]) -> Optional[ListNode]:

        dummy_move = ListNode(0)
        current = dummy_move 
        carry = 0 

        point1 = l1 
        point2 = l2 

        while point1 or point2 or carry:
            x1 = point1 if point1 else 0 
            x2 = point2 if point2 else 0 

            total = x1 + x2 + carry 
            carry = total//carry 
            unit = total % 10 

            current.next = ListNode(unit)
            current = current.next 

            if point1: 
                point1 = point1.next 
            
            if point2: 
                point2 = point2.next 
            
        return dummy_move.next 
```

## Three sum 

### Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.


```python
class ThreeSum:
    def __init__(self, nums:List[int]) -> Optional[List]:
        #first we sort the array
        nums.sort() 
        #define the output data structure 
        output = [] 
        
        n = len(nums)
        for i in range(n-2):
            if i > 0 and nums[i] == nums [i-1]:
                continue  ##if the current number and the next number are same skip them 
            
            ##edge case here 
            if nums[i] > 0 :
                break 
                
            #now that we have assigned a initial loop starting point here ,  assign the pointers pull as well
            l = i + 1 
            r = n - 1

            while l < r : 
                s = nums[i] + nums[l] + nums[r]

                if s == 0: 
                    output.append([nums[i],nums[l],nums[r]])

                    l += 1 
                    r -= 1 


                    while l < r and nums[l] == nums[l-1]:
                        l += 1 
                    
                    while l < r and nums[r] == nums[r-1]:
                         r -= 1 
                        
                elif s < 0:
                    l += 1 
                
                else :
                    r -= 1 
                
            
            return output

```

## Four Sum 

### Given an array nums of n integers, return an array of all the unique quadruplets [nums[a], nums[b], nums[c], nums[d]] such that:

0 <= a, b, c, d < n
a, b, c, and d are distinct.
nums[a] + nums[b] + nums[c] + nums[d] == target
You may return the answer in any order.



```python
class FourSum: 
    def __init__(self, nums:List[int], target:int)->List[int]:
        
        nums.sort()
        output = [] 
        n = len(nums)

        ## we can approach this problem in two pointer and two sum fashion 

        for i in range(n-3):
            if i > 0 and nums[i] == nums[i-1]:
                continue 
            
            min1 = nums[i]+nums[i+1]+nums[i+2]+nums[i+3]
            if min1 > target:
                break 

            max1 = nums[i]+nums[n-1]+nums[n-3]+nums[n-2]
            if max1 > target:
                continue 

            for j in range(i+1, n-2):
                if j < i+1 and nums[j] == nums[j-1]:
                    continue 

                min2 = nums[j]+nums[j+1]+nums[j+2]
                if min2 > target:
                    break 

                max2 = nums[j]+nums[n-1]+nums[n-2]
                if max2 > target:
                    continue  


                #now assigning the two pointer methods 
                l = j+1 
                r = n-1 

                while l < r:
                    total = nums[i]+nums[j]+nums[l]+nums[r]

                    # three edge cases , either total agrees with the target value , or if the total may be less than target value else if the total may be more than target value 
                    if total == target:
                        output.append([nums[i]+nums[j]+nums[l]+nums[r]])       ## our destined four sum array 

                        l += 1 
                        r -= 1 

                        #now that we moved one pointer to the left and one to the right 
                        ## we are checking for the duplicates here
                        while l < r and nums[l] == nums[l-1]: #if the current pointer value is same as the previous pointer
                            l += 1  # move one ahead

                        while l < r and nums[r] == nums[r+1]: #if the current right pointer is same as the previous pointer
                            r -= 1 #move one ahead 

                    elif total < target: #if the total value is too small for the target then, move the middle pointer ( left ) to one ahead, since the right pointer is at the end 
                         l += 1 

                    else: 
                        r -= 1  #or if the total value is too large for the the target value , move the highest value to drop one element 


        return output       # and the final output 
           

```

#  Reverse the Integer

### Given a signed 32-bit integer x, return x with its digits reversed. If reversing x causes the value to go outside the signed 32-bit integer range [-231, 231 - 1], then return 0.

Assume the environment does not allow you to store 64-bit integers (signed or unsigned).

 

Example 1:

Input: x = 123
Output: 321
Example 2:

Input: x = -123
Output: -321
Example 3:

Input: x = 120
Output: 21

```python
class ReverseInt:
    def __init__(self, x:int) -> int:
        x = str(x)

        is_negative = False 

        if x[0] == "-":
            is_negative = True 
            x = x[1:]
        x = int(x[::-1]) ##this way the str is now reversed and this is now converted to the integer

        if is_negative: 
            x = 0 - x
        
        if x > 2**31 -1 or x < 2 **31 : 
            return 0 
        
        else:
            return x 

```
